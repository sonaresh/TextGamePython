```python
"""
The following game was inspired by real treasures discovered in June 2011 in Thiruvananthapura, Kerala, India in Padmanabhaswamy temple.
The temple, which was established in the sixth century A.D. and has six vaults full of treasure, is one of the worldâ€™s richest. 
It has six vaults called Vault A, Vault B, Vault C, Vault D, Vault E, and Vault F. These vaults, also known as chambers, 
are crammed with treasure worth billions of dollars.The mysterious Vault B There are various more mythological conceptions associated with this vault, 
such as the vault is protected by guarded snakes, any human attempt to gain entry without using the Garuda Mantras, 
they would unleash great disasters upon the world, Vault B has still not been unlocked. 
It is predicted that the temple's Vault B would hold a considerable amount of treasures in the form of gold and other priceless things.

#Reference Pages:
https://www.indiatoday.in/magazine/special-report/story/20110718-who-owns-sree-padmanabhaswamy-temple-in-kerala-746812-2011-07-09
https://tinyurl.com/mfbm9c3e

"""


# importing exit form sys module
from sys import exit
# Importing random module
import random


# Name function 
def name_input():
    # Prompt message for name input
    prompt = "What is your name? :--  "
    # Initilizing name variable outside scope
    nameInput = None
    # While statement for valid name input
    while True:
        # Name input
        nameInput = input(prompt).title().strip()
        # Condition for valid name in string
        if len(nameInput) > 0 and not nameInput.isnumeric():
            # If name is valid then break the while loop
            break
        else:
            # If name is invalid then retry
            print("Enter your valid name")
    
    # Return name input
    return nameInput

def password_input():
    # Prompt message for name password input
    prompt = "Enter your password :  "
    # Initilizing variable outside scope
    passwordInput = None
    # While statement for valid input
    while True:
        # password input
        passwordInput = input(prompt).title().strip()
        # Condition for valid length in string
        if len(passwordInput) > 0:
            # If valid then break the while loop
            break
        else:
            # If invalid then retry
            print("Enter your password ")
    
    prompt2 = "confirm your password :  "
    # Initilizing variable outside scope
    password2Input = None
    # While statement for valid input
    while True:
        # password input
        password2Input = input(prompt2).title().strip()
        # Condition for valid password
        if password2Input == passwordInput:
            # If valid then break the while loop
            break
        else:
            # If invalid then retry
            print("Retry confirm password ")
    
    # Return name input
    return passwordInput

# Age function
def age_input():
    # Prompt message for age input
    prompt = "What is your age? :--  "
    # Initilizing age variable outside scope
    ageInput = None
    # While statement for valid age input
    while True:
        # Age input in integer
        ageInput = input(prompt).title()
        # Condition for valid age isnumeric
        if ageInput.isnumeric():
            # If age is valid then break the while loop
            break
        else:
            # If age is invalid then retry
            print("Enter your valid age in numeric")
    # Return age input
    return ageInput


# Assigining user name function to variable
user_name = name_input()

# Assigining user age function to variable
user_age = age_input()

password = password_input()


# Generate random user ID
def generate_random_user_id():
    # Replace spaces in user name
    name = user_name.replace(" ", "-")
    # Generate some random number between 10000 to 99999
    randomNumber = random.randint(10000, 99999)
    # Return user id
    user_id = f"{name}-{randomNumber}"
    return user_id


print(f"Hello, {user_name}. Let us go on a quest!")

# Assigining user age function to variable
userID = generate_random_user_id()

# Print user details
print(
    f'''
     ******** User details ********
     Name    : {user_name}
     Age     : {user_age}
     User ID : {userID}
     ******************************
    '''
)

# To terminate the process
def terminate():
    print('Processes terminated')
    # Execting process
    exit(0)

# Retry function
def retry_game():
    # Retry game prompt message 
    prompt = "Would you like to re-play the game and find the treasure? yes/no:--- "
    val = input(prompt).lower()
    # response function
    get_responce(val)


# Vaults and messages
vaults = {
    "vault1": {
        "msg": f"Congratulations {user_name} for finding the treasure;\n\thowever, this location has been closed for 130 years.\n\tOld crowns encrusted with diamonds and gems are among the treasures. 2,000 gold chains studded with diamonds,\n\tone of which is 18 feet long and weighs 10 kilograms. $9.3 Billion is the estimated worth.",
    },
    "vault2": {
        "msg": f"You've been killed by a deadly snake. Goodbye, {user_name}.\n",
    },
    "vault3": {
        "msg": f"Congratulations {user_name} on discovered the treasure;\n\twhich is worth a total of $19.4 million.\n\t450 gold pots, gold rods, gold umbrellas, daily puja ornaments, diamond necklaces,\n\tgold Garuda statue, diamond statue diamond studded crown, gold conch, gold bells, and gold flower vases.",
    },
    "vault4": {
        "msg": f"Congratulations {user_name} on finding the hidden treasure;\n\tThe set includes a diamond-studded crown and seven diamond bracelets.\nAmong the 400 gold scrappily (arrow tail-shaped) chains are gold incense stick stands, gold bows, and gold snake heads.\nTotal worth is $46 Millian",
    }
}


print('''
        One of the world's richest temples has five vaults labelled as Vault 1 to Vault 5.
        The treasure in some of these vaults (chambers) is worth billions of dollars.
        Are you ready to discover which vault contains treasure?
        \033[1mWarning: Some vaults are guarded by Giant Cobras, which can be deadly if you go too close.\033[0m
''')

# Yes or no response function
def get_responce(response):
    # This will check wehere response is valid
    if response in responses:
        print(responses[response]["msg"])
        # Execute operation function
        # Defined in response object
        responses[response]["operation"]()
    else:
        print("Enter valid option   yes/no")
        choose_yes_no()

# Choosing vaults
def choose_yes_no():
    # User choise prompt message
    prompt = "Are you certain you'll be able to locate the treasure? yes/no:--- "
    # Choosing yes or no input
    response = input(prompt).lower()
    
    # Calling response with input value
    get_responce(response)

    # if response is yes then print below
    if response == "yes":
        print('''
        \033[1mvault1:\033[0m More caves will appear along your path. seriously risky paths and a very dark place.
        \033[1mvault2:\033[0m You are entering a dark area with broken bridges, which might be risky.
        \033[1mvault3:\033[0m To get to the vault, you must swim in drained water, which contains many deadly insects,\n\tSome of which are poisonous.
        \033[1mvault4:\033[0m You may have to crawl in a small spot down an extremely narrow path.
        ==================================================================================================
        ''')


# Avaliable tools
# Along with operation function wich will decide process
tools = {
    "toolkit": {"msg": "You have been killed by Snake", "operation": retry_game},
    "key":  {"msg": "SUCCESS",  "operation": 'SUCCESS'},
    "grenade":   {"msg": "KILLED by an explosion", "operation": retry_game},
    "powertools": {"msg": "You'll need extra equipment to go treasure hunting.", "operation": choose_yes_no}
}

# Selecting tool
def choose_tools():
    # Select tools
    # Get all only keys from tools object
    tools_keys = [*tools]
    # Tools prompt message
    prompt = f"Select you requird tool from the required list {[*tools_keys]}\n"
    # select input of vaults from tools_key array
    selectTool = input(prompt).lower()
    if selectTool in tools_keys:
        selected_tool = tools[selectTool]
        print(f"selected {selectTool}")
        print(selected_tool["msg"])
        # Getting operation function from tool
        opertaion = selected_tool["operation"]
        # if operation is callable then which falles in this condition
        if callable(opertaion):
            # Execution the operation
            opertaion()
            # returining false for further execution
            return False
        # if operation is not callable this condition will accept
        else:
            # returining false for further process
            return True
    else:
        print("I didn't understand that could you please select item from given list")
        # Retry game if wrong input
        choose_tools()

# Selecting vault
def choose_vault(vaults_avaliable):
    # Choose Vault prompt message
    prompt = f"which vault would you like to enter? {vaults_avaliable}:--- "
    # Choose vaults from avaliable 
    treasure_val = input(prompt).lower()
    # if treasure_val contains in avaliable value
    if treasure_val in vaults_avaliable:
        # Assainging a treasure_val to variable
        vault = vaults[treasure_val]
        # Assaining choose_tools to variable
        res_tools = choose_tools()
        if res_tools == True:
            print(vault["msg"])
            # Retry function
            retry_game()

    else:
        print("I didn't understand that could you please try yes or no\n")
        # Again choose yes or no
        choose_yes_no()

# Selecting zone
def choose_zones():
    # zones object
    zones = {
        "EAST": ["vault1", "vault2"], "WEST": ["vault2", "vault3"],
        "NORTH": ["vault3", "vault4"], "SOUTH": ["vault4", "vault1"]
    }
    # Get zones keys from object
    zones_keys = [*zones]
    print(f'Avaliable zones {zones_keys}')
    print('Selecting random zone')
    # Selecting random choise from zone_keys
    random_zone = random.choice(zones_keys)
    print(f"Your at {random_zone} zone")
    print(f"Your are near to {zones[random_zone]}")
    # chosing vault from selected zone
    choose_vault(zones[random_zone])

# Response variables
responses = {
    "yes": {
        "msg": '''
        You've arrived in Mystirius, a place where many people were looking for treasure.
        They all died in the end, so select your way of treasure hunting carefully.
        ==================================================================================================
        ''',
        "operation": choose_zones
    },
    "no": {
        "msg": f"You are not ready for Treasure hunting. Goodbye, {user_name} .",
        "operation": terminate
    }
}

# Choosing yes or no function
choose_yes_no()
```
